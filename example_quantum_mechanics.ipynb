# Quantum Computing - Quantum Mechanics

### 0 - Setup

```python
import qiskit as q
from qiskit.visualization import plot_histogram, plot_state_qsphere

import warnings
warnings.filterwarnings("ignore")
```

### 1 - Single Qubit

###### Initialize

```python
# model
c = q.QuantumCircuit(1)

# run
backend = q.Aer.get_backend('statevector_simulator')
res = q.execute(c, backend).result()

# plot
vec = res.get_statevector()
plot_state_qsphere(vec, figsize=(5,5))
```

###### Superposition

```python
# model
c = q.QuantumCircuit(1)
c.h(0) #<-- Hadamard gate

# run
backend = q.Aer.get_backend('statevector_simulator')
res = q.execute(c, backend).result()

# plot
vec = res.get_statevector()
plot_state_qsphere(vec, figsize=(5,5))
```

###### Measurement

```python
# model
c = q.QuantumCircuit(1,1)
c.h(0) #<-- Hadamard gate
c.measure(qubit=0, cbit=0)

# run
backend = q.Aer.get_backend('statevector_simulator')
res = q.execute(c, backend).result()

# plot
vec = res.get_statevector()
plot_state_qsphere(vec, figsize=(5,5))
```

### 2 - Circuits

###### Single qubit

```python
# model
c = q.QuantumCircuit(1,1, name="single qubit")
c.h(0) #<-- Hadamard gate
c.measure(qubit=0, cbit=0)

# plot
c.draw(output="mpl", scale=0.7)
```

```python
# run
backend = q.Aer.get_backend('qasm_simulator')
res = q.execute(c, backend, shots=1000).result()

# plot
dist = res.get_counts(c)
plot_histogram(dist, figsize=(4,2), color="black", title=c.name)
```

###### Multi qubits

```python
# model
c = q.QuantumCircuit(2,2, name="multi qubits")
c.h([0,1]) #<-- Hadamard gate
c.cx(0,1) #<-- CNOT gate
c.measure(qubit=[0,1], cbit=[0,1])

# plot
c.draw(output="mpl", scale=0.7)
```

```python
# run
backend = q.Aer.get_backend('qasm_simulator')
res = q.execute(c, backend, shots=1000).result()

# plot
dist = res.get_counts(c)
plot_histogram(dist, figsize=(4,2), color="black", title=c.name)
```
